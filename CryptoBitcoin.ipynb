# Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Normalize the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)

# Display the first 5 rows
df_scaled.head()
```

Create Technical Indicators
1ï¸âƒ£ Simple Moving Average (SMA) & Exponential Moving Average (EMA)

```python
# 10-day & 30-day Simple Moving Averages
df_scaled['SMA_10'] = df_scaled['Close'].rolling(window=10).mean()
df_scaled['SMA_30'] = df_scaled['Close'].rolling(window=30).mean()

# 10-day & 30-day Exponential Moving Averages
df_scaled['EMA_10'] = df_scaled['Close'].ewm(span=10, adjust=False).mean()
df_scaled['EMA_30'] = df_scaled['Close'].ewm(span=30, adjust=False).mean()
```

2ï¸âƒ£ Relative Strength Index (RSI)

```python
def compute_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df_scaled['RSI'] = compute_rsi(df_scaled['Close'])
```

3ï¸âƒ£ Moving Average Convergence Divergence (MACD)

```python
# MACD Line
df_scaled['MACD'] = df_scaled['EMA_10'] - df_scaled['EMA_30']

# MACD Signal Line (9-day EMA of MACD)
df_scaled['MACD_Signal'] = df_scaled['MACD'].ewm(span=9, adjust=False).mean()
```

```python
df_scaled['MACD_Signal'] 
```

# Train-Test Split & Data Preparation

```python
df_scaled
```

```python
# Drop NaN values from feature engineering
df_scaled.dropna(inplace=True)

# Define X (features) & Y (target - 'Close' price)
X = df_scaled.drop(columns=['Close'])  # All features except Close
Y = df_scaled['Close']  # Close price as target

# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=False)

print(f"Train Data Shape: {X_train.shape}, Test Data Shape: {X_test.shape}")
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Train Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)

# Predictions
lr_pred = lr_model.predict(X_test)

# Evaluate Linear Regression
lr_mae = mean_absolute_error(Y_test, lr_pred)
lr_rmse = np.sqrt(mean_squared_error(Y_test, lr_pred))
lr_r2 = r2_score(Y_test, lr_pred)

print(f"Linear Regression -> MAE: {lr_mae}, RMSE: {lr_rmse}, R2: {lr_r2}")
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(14,6))
plt.plot(Y_test.values, label="Actual Prices", color='blue')
plt.plot(lr_pred, label="Predicted Prices", color='red', linestyle='dashed')
plt.legend()
plt.title("Bitcoin Price Prediction - Actual vs Predicted")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.show()
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

# Predictions
rf_pred = rf_model.predict(X_test)

# Evaluate Random Forest
rf_mae = mean_absolute_error(Y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(Y_test, rf_pred))
rf_r2 = r2_score(Y_test, rf_pred)

print(f"Random Forest -> MAE: {rf_mae}, RMSE: {rf_rmse}, R2: {rf_r2}")
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(14,6))
plt.plot(Y_test.values, label="Actual Prices", color='blue')
plt.plot(rf_pred, label="Predicted Prices", color='red', linestyle='dashed')
plt.legend()
plt.title("Bitcoin Price Prediction - Actual vs Predicted")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.show()
```

# ARIMA for Time-Series Forecasting

```python
from statsmodels.tsa.arima.model import ARIMA

# Train ARIMA Model
arima_model = ARIMA(Y_train, order=(5,1,0))  # ARIMA (p,d,q)
arima_model_fit = arima_model.fit()

# Predictions
arima_pred = arima_model_fit.forecast(steps=len(Y_test))

# Evaluate ARIMA
arima_mae = mean_absolute_error(Y_test, arima_pred)
arima_rmse = np.sqrt(mean_squared_error(Y_test, arima_pred))
arima_r2 = r2_score(Y_test, arima_pred)

print(f"ARIMA -> MAE: {arima_mae}, RMSE: {arima_rmse}, R2: {arima_r2}")
```

```python
# Compare all models
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "ARIMA"],
    "MAE": [lr_mae, rf_mae,  arima_mae],
    "RMSE": [lr_rmse, rf_rmse,  arima_rmse],
    "R2 Score": [lr_r2, rf_r2,  arima_r2]
})

print(results)

# Plot Actual vs Predicted for the best model
plt.figure(figsize=(12,6))
plt.plot(Y_test.values, label="Actual Prices", color='blue')
plt.plot(rf_pred, label="Predicted Prices (Best Model)", color='red')
plt.legend()
plt.title("Bitcoin Price Prediction")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential, model_from_json
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping



ds['Close'] = ds['Close'].astype(float)  # Ensure float type

# Scale the 'Close' prices
sc = MinMaxScaler()
scaled_data = sc.fit_transform(ds['Close'].values.reshape(-1, 1))

# Function to prepare LSTM input data
def prepare_data(data, past_days):
    X, Y = [], []
    for i in range(len(data) - past_days):
        X.append(data[i:i + past_days])
        Y.append(data[i + past_days])
    return np.array(X), np.array(Y)

# Define past days to consider
past_days = 30
X, Y = prepare_data(scaled_data, past_days)

# Split data into training and testing sets (80% train, 20% test)
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]

# Reshape data for LSTM input
X_train_lstm = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test_lstm = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build the LSTM Model
print("Building LSTM model...")
lstm_model = Sequential([
    LSTM(64, activation='relu', return_sequences=True, input_shape=(past_days, 1)),
    Dropout(0.2),
    LSTM(64, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

# Compile the model
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Define Early Stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the Model
lstm_model.fit(X_train_lstm, Y_train, epochs=30, batch_size=32, validation_data=(X_test_lstm, Y_test), verbose=1, callbacks=[early_stopping])

# Save Model Architecture and Weights
print("Saving model...")
model_json = lstm_model.to_json()
with open("lstm_model.json", "w") as json_file:
    json_file.write(model_json)

lstm_model.save_weights("lstm_weights.h5")
print("Model saved successfully!")

# Load the Model & Weights for Future Use
print("Loading saved model...")
with open("lstm_model.json", "r") as json_file:
    loaded_model_json = json_file.read()
loaded_model = model_from_json(loaded_model_json)
loaded_model.load_weights("lstm_weights.h5")

# Compile loaded model
loaded_model.compile(optimizer='adam', loss='mean_squared_error')
print("Model loaded successfully!")

# Make Predictions
predictions = loaded_model.predict(X_test_lstm)

# Convert predictions back to original scale
predictions = sc.inverse_transform(predictions)

# Visualization of Results
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(ds.index[-len(Y_test):], sc.inverse_transform(Y_test.reshape(-1,1)), label="Actual Price")
plt.plot(ds.index[-len(Y_test):], predictions, label="Predicted Price", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.show()
```
